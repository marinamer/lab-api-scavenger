# Challenge 1: Fork Languages

```python
import requests
import json
import pandas as pd
from requests.auth import HTTPBasicAuth
from pandas.io.json import json_normalize
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
url = 'https://api.github.com/repos/ironhack-datalabs/datalis0819/forks'

response = requests.get(url,auth = HTTPBasicAuth('marinamer', 'TOKEN'))
results = response.json()
results
```

```python
flattened_data = pd.json_normalize(results)
language = [i for i in flattened_data['language']]
language = set(language)
language
```
